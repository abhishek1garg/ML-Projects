```python
import glob
import pandas as pd
```

```python
path = r'' # use your path
all_files = glob.glob("*.csv")

li = []

print("There are "+str(len(all_files))+" files")
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

allstations = pd.concat(li, axis=0, ignore_index=True)
```

```python
allstations.drop(columns=["Unnamed: 0"]).to_csv("./New/AQI_Data_27Jan2021.csv",encoding="utf-8")
```
