```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time
from datetime import datetime,timedelta
```

```python
browser=webdriver.Chrome()
browser.get("https://app.cpcbccr.com/AQI_India/")
```

```python
browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
statesDropdown=browser.find_element_by_id("states")
statesOptions=statesDropdown.find_elements_by_tag_name("option")
statesList=[stateOption.get_attribute('value') for stateOption in statesOptions if stateOption.get_attribute('value')!="-1"]
print(statesList)
citiesDropdown=browser.find_element_by_id("cities")
citiesOptions=citiesDropdown.find_elements_by_tag_name("option")
stationsDropdown=browser.find_element_by_id("stations")
```

```python
AQIDict=dict()
stationDictList=[]
for state in statesList:
    Select(statesDropdown).select_by_value(state)
    #time.sleep(2)
    citiesList=[cityOption.get_attribute('value') for cityOption in citiesOptions if cityOption.get_attribute('value')!="-1"]
    print(citiesList)
    for city in citiesList:
        Select(citiesDropdown).select_by_value(city)
        stationsOptions=stationsDropdown.find_elements_by_tag_name("option")
        stationsList=[stationOption.get_attribute('value') for stationOption in stationsOptions if stationOption.get_attribute('value')!="-1"]
        stationNamesList=[stationOption.get_attribute('text') for stationOption in stationsOptions if stationOption.get_attribute('value')!="-1"]
        #print(len(stationsList))
        #print(stationNamesList)
        for index,txt in enumerate(stationsList):
            stationDict={"State":state,
                         "City":city,
                         "Station_ID":txt,
                         "Station_Name":stationNamesList[index]}
            stationDictList.append(stationDict)
```

```python
row_list=[]
timeInputText.clear()
timeInputText.send_keys("0:00")
time.sleep(1)
#timeInputText.send_keys(Keys.ENTER)
prevStation=majorcitystatonsAQIdf.loc[111,"Station_ID"]
prevCity=majorcitystatonsAQIdf.loc[111,"City"]
for index, row in majorcitystatonsAQIdf.iterrows():
    Select(statesDropdown).select_by_value("-1")
    Select(statesDropdown).select_by_value(row["State"])
    Select(citiesDropdown).select_by_value("-1")
    Select(citiesDropdown).select_by_value(row["City"])
    Select(stationsDropdown).select_by_value("-1")
    Select(stationsDropdown).select_by_value(row["Station_ID"])
    dateInputText.clear()
    dateInputText.send_keys(row["Date"])
    dateInputText.send_keys(Keys.ENTER)
    time.sleep(1)
    if(len(browser.find_elements_by_xpath("/html/body/div[3]/div[1]/div[2]/div[2]/div[2]/div[3]/div[1]/table/tbody"))>0):
        print("Found AQI for "+str(row["State"])+" - "+str(row["City"])+" - "+str(row["Station_Name"])+" - "+row["Date"])
        aqiWebTable=browser.find_element_by_xpath("/html/body/div[3]/div[1]/div[2]/div[2]/div[2]/div[3]/div[1]/table/tbody").get_attribute("innerHTML")
        soup = BeautifulSoup(aqiWebTable, "lxml")
        element_name_cols=soup.find_all("td",{'class':'element-name'})
        element_names=[col.text for col in element_name_cols]
        #print(element_names)
        aqi_value_avg_cols=soup.find_all("td",{'class':'avg-value'})
        aqi_avg_values=[col.text for col in aqi_value_avg_cols]
        aqi_value_min_cols=soup.find_all("td",{'class':'min-value'})
        aqi_min_values=[col.text for col in aqi_value_min_cols]
        aqi_value_max_cols=soup.find_all("td",{'class':'max-value'})
        aqi_max_values=[col.text for col in aqi_value_max_cols]
        for indexlist, val in enumerate(element_names):
            majorcitystatonsAQIdf.at[index,val+"_average"]=aqi_avg_values[indexlist]
            majorcitystatonsAQIdf.at[index,val+"_minimum"]=aqi_min_values[indexlist]
            majorcitystatonsAQIdf.at[index,val+"_maximum"]=aqi_max_values[indexlist]
            #print(aqi_avg_values[indexlist])
            #print(aqi_min_values[indexlist])
            #print(aqi_max_values[indexlist])
            time.sleep(1)
        #print(majorcitystatonsAQIdf.iloc[index])
    if prevStation!=row["Station_ID"]:
        majorcitystatonsAQIdf[majorcitystatonsAQIdf["Station_ID"]==prevStation].to_csv("AQI_Data_"+prevCity+"_"+prevStation+".csv")
        prevStation=row["Station_ID"]
        prevCity=row["City"]
```
